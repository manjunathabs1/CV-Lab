```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = r'C:\Users\Siddaraju.B.R\Downloads\th.jpeg'  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Apply the Laplacian operator
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# Convert to 8-bit absolute value for visualization
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Display the Laplacian image
plt.subplot(1, 2, 2)
plt.imshow(laplacian_abs, cmap='gray')
plt.title('Laplacian Image')
plt.axis('off')

plt.tight_layout()
plt.show()

# Save the Laplacian image (optional)
output_path = 'laplacian_image.jpg'
cv2.imwrite(output_path, laplacian_abs)
```
