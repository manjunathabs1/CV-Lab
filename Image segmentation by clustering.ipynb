```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io

# Load the image
image_path = r'C:\Users\Siddaraju.B.R\Downloads\th.jpeg'  # Replace with the path to your image
image = io.imread(image_path)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Reshape the image into a 2D array of pixels
pixels = image.reshape(-1, 3)

# Normalize pixel values (optional but improves clustering)
pixels = pixels / 255.0

# Apply K-Means clustering
num_clusters = 5  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(pixels)
segmented_pixels = kmeans.cluster_centers_[kmeans.labels_]

# Reshape back to the original image shape
segmented_image = segmented_pixels.reshape(image.shape)

# Display the segmented image
plt.imshow(segmented_image)
plt.title(f'Segmented Image with {num_clusters} Clusters')
plt.axis('off')
plt.show()

# Save the segmented image (optional)
output_path = 'segmented_image.jpg'
io.imsave(output_path, (segmented_image * 255).astype(np.uint8))
```
